```python
from fatiando import gridder, utils, gravmag, mesher
from fatiando.vis import mpl

x, y, z = gridder.scatter([0, 10000, 0, 20000], 400, z=-1)
spheres = [mesher.Sphere(5000, 10000, 1000, radius=500, props={'density':1000})]
gz = gravmag.sphere.gz(x, y, z, spheres)
mpl.figure()
mpl.axis('scaled')
mpl.plot(y, x, '.k')
mpl.contourf(y, x, gz, (50, 50), 40, interp=True)
mpl.m2km()
mpl.savefig('grav_example.png')
```
